```python
import matplotlib.pyplot as plt
import numpy as np

# Simulated data for EDA
np.random.seed(42)

# Angle ranges for exercises
bicep_curl_angles = np.random.uniform(30, 160, 100)
shoulder_press_angles = np.random.uniform(60, 160, 100)
lateral_raise_angles = np.random.uniform(30, 90, 100)

# Repetition counts (simulate rep counts for each exercise)
bicep_curl_reps = np.random.randint(8, 15, 100)
shoulder_press_reps = np.random.randint(6, 12, 100)
lateral_raise_reps = np.random.randint(10, 20, 100)

# Stage transition counts (simulate transitions for each exercise)
stage_transitions_bicep = np.random.randint(1, 2, 100)
stage_transitions_shoulder_press = np.random.randint(1, 2, 100)
stage_transitions_lateral_raise = np.random.randint(1, 2, 100)

# Plot angle distribution for each exercise
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].hist(bicep_curl_angles, bins=20, color='skyblue', edgecolor='black')
axes[0].set_title('Bicep Curl Angle Distribution')
axes[0].set_xlabel('Angle (degrees)')
axes[0].set_ylabel('Frequency')

axes[1].hist(shoulder_press_angles, bins=20, color='salmon', edgecolor='black')
axes[1].set_title('Shoulder Press Angle Distribution')
axes[1].set_xlabel('Angle (degrees)')
axes[1].set_ylabel('Frequency')

axes[2].hist(lateral_raise_angles, bins=20, color='lightgreen', edgecolor='black')
axes[2].set_title('Lateral Raise Angle Distribution')
axes[2].set_xlabel('Angle (degrees)')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

# Plot rep counts for each exercise
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].hist(bicep_curl_reps, bins=10, color='skyblue', edgecolor='black')
axes[0].set_title('Bicep Curl Rep Count')
axes[0].set_xlabel('Reps')
axes[0].set_ylabel('Frequency')

axes[1].hist(shoulder_press_reps, bins=10, color='salmon', edgecolor='black')
axes[1].set_title('Shoulder Press Rep Count')
axes[1].set_xlabel('Reps')
axes[1].set_ylabel('Frequency')

axes[2].hist(lateral_raise_reps, bins=10, color='lightgreen', edgecolor='black')
axes[2].set_title('Lateral Raise Rep Count')
axes[2].set_xlabel('Reps')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

# Plot stage transitions for each exercise
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].hist(stage_transitions_bicep, bins=3, color='skyblue', edgecolor='black')
axes[0].set_title('Bicep Curl Stage Transitions')
axes[0].set_xlabel('Transitions')
axes[0].set_ylabel('Frequency')

axes[1].hist(stage_transitions_shoulder_press, bins=3, color='salmon', edgecolor='black')
axes[1].set_title('Shoulder Press Stage Transitions')
axes[1].set_xlabel('Transitions')
axes[1].set_ylabel('Frequency')

axes[2].hist(stage_transitions_lateral_raise, bins=3, color='lightgreen', edgecolor='black')
axes[2].set_title('Lateral Raise Stage Transitions')
axes[2].set_xlabel('Transitions')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
